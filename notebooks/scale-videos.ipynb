```python
import os
import numpy as pd
import cv2
import matplotlib.pyplot as plt
from moviepy.editor import VideoFileClip
%matplotlib inline
plt.rcParams['figure.facecolor'] = 'white'
```

# Lens D-high HD

```python
vid_filename = "D_high_HD_2024_0216_141107_006"
vid_extension = ".MP4"
vid_filepath = f"../data/videos/scale-videos/{vid_filename}{vid_extension}"

fps = VideoFileClip(vid_filepath).fps  # get from file original, not segmented
resolution = VideoFileClip(vid_filepath).size

print(fps, resolution)
```

```python
# read in image
video = cv2.VideoCapture(vid_filepath)
_, frame = video.read()

plt.figure()
plt.imshow(frame)
plt.plot([0, 1200],[400, 400], "or")
plt.minorticks_on()
```

```python
pixel_width = 1200 - 0  # from left to right
real_width = 1000  # in micrometer

real_width_per_pixel = real_width / pixel_width
pixels_per_real_width = pixel_width / real_width
print(real_width_per_pixel, pixels_per_real_width)

desired_scale_width = 250  # in mu
scaled_width = desired_scale_width * pixels_per_real_width
print(scaled_width)

scalebar_posx = 400
scalebar_posy = 400

plt.figure()
plt.imshow(frame)
plt.plot([scalebar_posx, scalebar_posx + scaled_width], [scalebar_posy, scalebar_posy], "-or")
plt.minorticks_on()
```
