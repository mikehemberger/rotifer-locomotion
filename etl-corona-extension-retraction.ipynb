```python
import os
import pandas as pd
from PIL import Image
import numpy as np
#import matplotlib as mpl
import cv2  # conda install -c conda-forge opencv
from moviepy.editor import VideoFileClip
from moviepy.editor import *


import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams['axes.grid'] = False
#plt.style.use('dark_background')
plt.rcParams["axes.spines.top"] = False
plt.rcParams["axes.spines.right"] = False
plt.rcParams["font.size"] = 15
plt.rcParams["savefig.bbox"] = "tight"
```

```python
vid_dir = "pond-water-rotifers"
vid_path = f"./data/videos/{vid_dir}/"


vids = ["2024_0126_162412_122.MP4", "2024_0126_162616_123.MP4", "2024_0126_162729_125.MP4"]
ts = [41,0,50]
te = [59,10,60]
#"2024_0126_162729_125.MP4" - 33-37 for retraction! fast
vids = ["2024_0126_162729_125.MP4"]
ts = [33]
te = [37]
for vid_filename, trim_start, trim_end in zip(vids, ts, te):
    trim_duration = trim_end - trim_start
    vid_filepath = f"{vid_path}{vid_filename}"

    PROCESS_VIDEO = True
    SAVE_VIDEO = True
    if PROCESS_VIDEO:
        video_clip = VideoFileClip(vid_filepath).subclip(trim_start, trim_end)
        duration = video_clip.duration
        resolution = video_clip.size
        fps = video_clip.fps

        print(duration, fps, resolution)
        if SAVE_VIDEO:
            video_clip.write_videofile(f"{vid_path}{vid_filename.split('.')[0]}_subclip_{trim_duration}sec_ts{trim_start}-te{trim_end}.mp4",
                                        fps=fps, codec="libx264", audio=False)
        video_clip.close()
```

```python
plt.rcParams['figure.facecolor'] = 'white'
for vid_filename, trim_start, trim_end in zip(vids, ts, te):
    trim_duration = trim_end - trim_start
    vid_filepath = f"{vid_path}{vid_filename}"

    SHOW_FRAME = True
    if SHOW_FRAME:
        vid_filepath_trim = f"{vid_path}{vid_filename.split('.')[0]}_subclip_{trim_duration}sec_ts{trim_start}-te{trim_end}.mp4"
        cap = cv2.VideoCapture(vid_filepath_trim)
        ret, frame = cap.read()
        height, width, _ = frame.shape
        print(height, width)

        plt.figure()
        plt.imshow(frame)
    #plt.imshow(frame[100:900, 1100:1500])

x1_ = [950, 950, 700]
x2_ = [1750, 1750, 1100]
y1_ = [500, 500, 350]
y2_ = [1400, 1400, 700]
x1_ = [700]
x2_ = [1200]
y1_ = [350]
y2_ = [700]
SAVE_VIDEO = True
if SAVE_VIDEO:
    for nth, (vid_filename, trim_start, trim_end) in enumerate(zip(vids, ts, te)):
        trim_duration = trim_end - trim_start
        vid_filepath_trim = f"{vid_path}{vid_filename.split('.')[0]}_subclip_{trim_duration}sec_ts{trim_start}-te{trim_end}.mp4"

        video_clip = VideoFileClip(vid_filepath_trim)
        fps = video_clip.fps
        video_crop = video_clip.fx(vfx.crop, x1=x1_[nth], x2=x2_[nth], y1=y1_[nth], y2=y2_[nth]) #
        video_crop.write_videofile(f"{vid_path}{vid_filename.split('.')[0]}_subclip_{trim_duration}sec_ts{trim_start}-te{trim_end}_cropped.mp4", fps=int(fps), codec="libx264", audio=False)
        video_crop.close()
        video_clip.close()
```
